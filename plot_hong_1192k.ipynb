```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
H2O_CSV = 'hong_1398K_h2o.csv'
OH_CSV  = 'hong_1398K_oh.csv'

# skipinitialspace strips the leading space in column names
df_h2o = pd.read_csv(H2O_CSV, skipinitialspace=True)
df_oh  = pd.read_csv(OH_CSV,  skipinitialspace=True)

print('H2O columns:', df_h2o.columns.tolist())
print('OH  columns:', df_oh.columns.tolist())
print(f'H2O: {len(df_h2o)} pts, t=[{df_h2o["Time [ms]"].min():.5f}, {df_h2o["Time [ms]"].max():.5f}] ms')
print(f'OH:  {len(df_oh)} pts,  t=[{df_oh["Time [ms]"].min():.5f}, {df_oh["Time [ms]"].max():.5f}] ms')
```

```python
fig, ax = plt.subplots(figsize=(7, 5))

ax.plot(df_h2o['Time [ms]'], df_h2o['[H2O] ppm'],
        'o', mfc='none', mec='steelblue', ms=5, alpha=0.7)

ax.set_xlabel('Time [ms]')
ax.set_ylabel('[H$_2$O] / ppm')
ax.set_xlim([0, 0.10])
ax.set_ylim([1000, 4000])
ax.set_title('H$_2$O — Hong et al. 1192 K')
ax.grid(True, ls='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 5))

ax.plot(df_oh['Time [ms]'], df_oh['[OH] ppm'],
        'o', mfc='none', mec='tomato', ms=5, alpha=0.7)

ax.set_xlabel('Time [ms]')
ax.set_ylabel('[OH] / ppm')
ax.set_title('OH — Hong et al. 1192 K')
ax.grid(True, ls='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))

# H2O
ax1.plot(df_h2o['Time [ms]'], df_h2o['[H2O] ppm'],
         'o', mfc='none', mec='steelblue', ms=5, alpha=0.7)
ax1.set_xlabel('Time [ms]')
ax1.set_ylabel('[H$_2$O] / ppm')
ax1.set_xlim([0, 0.10])
ax1.set_ylim([1000, 4000])
ax1.set_title('H$_2$O')
ax1.grid(True, ls='--', alpha=0.4)

# OH
ax2.plot(df_oh['Time [ms]'], df_oh['[OH] ppm'],
         'o', mfc='none', mec='tomato', ms=5, alpha=0.7)
ax2.set_xlabel('Time [ms]')
ax2.set_ylabel('[OH] / ppm')
ax2.set_title('OH')
ax2.grid(True, ls='--', alpha=0.4)

plt.suptitle('Hong et al. 1192 K experimental data', fontsize=12, fontweight='bold')
plt.tight_layout()
plt.show()
```
